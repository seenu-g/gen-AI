```python
from transformers import pipeline
```

```python
sentiment_analyst = pipeline("sentiment-analysis")
```

```python
sentiment_analyst("I've been waiting for a HuggingFace course my whole life.")

sentiment_analyst(
    ["I've been waiting for a HuggingFace course my whole life.", "I hate this so much!"]
)
```

```python
classifier = pipeline("zero-shot-classification")
```

```python
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
# classify texts that haven’t been labelled. This is a common scenario in real-world projects because annotating text is usually time-consuming and requires domain expertise.
# classify the text using any other set of labels you like
```

```python
text_generator = pipeline("text-generation")
```

```python
text_generator("In this course, we will teach you how to")
```

```python
text_generator = pipeline("text-generation", model="distilgpt2")
```

```python
text_generator(
    "In this course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
# generate text in another language.
```

```python
unmasker = pipeline("fill-mask")
```

```python
unmasker("This course will teach you all about <mask> models.", top_k=2)
#  fill in the blanks in a given text:
# how many possibilities you want to be displayed.
```

```python
question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn",
)
# pipeline works by extracting information from the provided context; it does not generate the answer.
```

```python
summarizer = pipeline("summarization")
summarizer(
    """
    America has changed dramatically during recent years. Not only has the number of
    graduates in traditional engineering disciplines such as mechanical, civil,
    electrical, chemical, and aeronautical engineering declined, but in most of
    the premier American universities engineering curricula now concentrate on
    and encourage largely the study of engineering science. As a result, there
    are declining offerings in engineering subjects dealing with infrastructure,
    the environment, and related issues, and greater concentration on high
    technology subjects, largely supporting increasingly complex scientific
    developments. While the latter is important, it should not be at the expense
    of more traditional engineering.

    Rapidly developing economies such as China and India, as well as other
    industrial countries in Europe and Asia, continue to encourage and advance
    the teaching of engineering. Both China and India, respectively, graduate
    six and eight times as many traditional engineers as does the United States.
    Other industrial countries at minimum maintain their output, while America
    suffers an increasingly serious decline in the number of engineering graduates
    and a lack of well-educated engineers.
"""
)

# reducing a text into a shorter text while keeping all (or most) of the important aspects referenced in the text.
```

```python
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-fr-en") # translating from French to English:
translator("Ce cours est produit par Hugging Face.")
```

```python
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fr") # translating from English to French
translator("this machine learning course was produced by Rohan")
```

```python
#  Fine-tuning the model on your data won’t make this intrinsic bias disappear.
unmasker = pipeline("fill-mask", model="bert-base-uncased")
result = unmasker("This man works as a [MASK].")
print([r["token_str"] for r in result])

result = unmasker("This woman works as a [MASK].")
print([r["token_str"] for r in result])
```

```python
# Loading a model within a pipeline is extremely simple.
# The only thing you need to watch out for is that the chosen checkpoint is suitable for the task it’s going to be used for
camembert_fill_mask = pipeline("fill-mask", model="camembert-base")
results = camembert_fill_mask("Le camembert est <mask> :)")
```

```python
from transformers import CamembertTokenizer, CamembertForMaskedLM
#instantiate the checkpoint using the model architecture directly:
tokenizer = CamembertTokenizer.from_pretrained("camembert-base")
model = CamembertForMaskedLM.from_pretrained("camembert-base")
```

```python
from transformers import AutoTokenizer, AutoModelForMaskedLM
# recommend using the Auto* classes instead, as these are by design architecture-agnostic
tokenizer = AutoTokenizer.from_pretrained("camembert-base")
model = AutoModelForMaskedLM.from_pretrained("camembert-base")
```
